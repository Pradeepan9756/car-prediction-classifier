```python
x = df.iloc[:,:-1]
x.head()
```

```python
y = df.iloc[:,-1:]
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size =0.27,random_state =42)
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(random_state =42)
model.fit(x_train,y_train)
```

```python
tr_score = model.score(x_train,y_train)
te_score = model.score(x_test,y_test)
print(te_score,tr_score)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.27)
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(random_state=42)
model.fit(x_train,y_train)



tr_train=model.score(x_train,y_train)
te_test=model.score(x_test,y_test)
print(tr_train,te_test)
```
